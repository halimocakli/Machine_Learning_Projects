```python
# Gerekli Python kütüphanelerini import edelim
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
X_train = np.array([[165, 68],
                    [173, 89],
                    [182, 82],
                    [190, 86],
                    [154, 52],
                    [161, 59],
                    [185, 72],
                    [157, 52],
                    [170, 67]])
```

```python
y_train = ["W", "M", "M", "M", "W", "W", "M", "W", "W"]
```

```python
plt.figure(figsize=(11, 6), dpi=150)
plt.title("Height and Weight Values in Regards to Genders")
plt.xlabel("Height - CM")
plt.ylabel("Weight - KG")

for i, x in enumerate(X_train):
    plt.scatter(x[0], x[1], c="k", marker="^" if y_train[i] == "M" else "v")

plt.grid(True)
plt.show()
```

```python
# Kategorik veri olarak yani harf olarak girilen cinsiyet etiketlerini (K - E) nümerik yani sayısal veri haline dönüştürelim

from sklearn.preprocessing import LabelBinarizer

label = LabelBinarizer()
y_train = label.fit_transform(y_train)
```
